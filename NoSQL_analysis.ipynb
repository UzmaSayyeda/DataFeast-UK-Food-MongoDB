### 1. Which establishments have a hygiene score equal to 20?

```python
# find the establishments with a hygiene score fo 20

[ x for x in establishments.find({"scores.Hygiene" : {"$gte":20}})]
```

```python
# Use count_documents to display the number of documents in the result

establishments.count_documents({"scores.Hygiene" : {"$gte":20}})
```

```python
# Display the first document in the results using pprint

pprint([ x for x in establishments.find({"scores.Hygiene" : {"$gte":20}})][1])
```

```python
# Convert the result to a Pandas DataFrame

df = pd.DataFrame([ x for x in establishments.\
                   find({"scores.Hygiene" : {"$gte":20}})])
```

```python
# display the number of rows in the Dataframe

print(f" The number of rows are",df.shape[0])
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has 
# a RatingValue greater than or equal to 4.

query = {"LocalAuthorityName": {"$regex":"London"},
                    "RatingValue" : {"$gte" : 4}}

results = [x for x in establishments.find(query)]
```

```python
# Use count_documents to display the number of documents in the result

print(f"the number of documents are",establishments.count_documents(query))
```

```python
# Display the first document in the results using pprint

pprint(results[1])
```

```python
# Convert the result to a Pandas DataFrame

df2 = pd.DataFrame(results)
```

```python
# Display the number of rows in the DataFrame

print(f"The number of rows in second dataframe are", df2.shape[0])
```

```python
# Display the first 10 rows of the DataFrame

df2.head(10)
```
